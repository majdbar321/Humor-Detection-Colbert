```python
# import necessary packages
import tensorflow as tf
from embeddings_extract import *
from embeddings_utils import *
from load_data import *
```

```python
# Load data and split into train and validation set
dataload = LoadData()
df = dataload.load_data("path_to_data")

X_train, X_val, y_train, y_val = dataload.split_data()


# Load BERT tokenizer and model
tokenizer, bert_model = load_bert('gaunernst/bert-small-uncased')

```

```python
# using MirroredStrategy to distribute the model across multiple GPUs
strategy = tf.distribute.MirroredStrategy()
```

```python
# Extract BERT embeddings for the training set and validation set on Words level
extract_words_embeddings(X_train,X_val,strategy, tokenizer, bert_model)

# Extract BERT embeddings for the training set and validation set on Sentences level
extract_sentences_embeddings(X_train,X_val,strategy, tokenizer, bert_model)
```
