```python
# Extract BERT embeddings for the training set and validation set on Words level
extract_words_embeddings(X_train,X_val,strategy, tokenizer, bert_model)

# Extract BERT embeddings for the training set and validation set on Sentences level
extract_sentences_embeddings(X_train,X_val,strategy, tokenizer, bert_model)
```
